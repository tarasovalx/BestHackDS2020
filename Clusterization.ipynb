```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

scaler = StandardScaler()
scaler.fit(df)
scaled_df = scaler.transform(df)
```

```python
df_scaled = pd.DataFrame(scaled_df)
df_scaled.columns = df.columns
df_scaled
```

```python
def threshold_scale(w):
    if(w > 1.25):
        return w ** 6
    return w

df_scaled['Water_(g)'] = df_scaled['Water_(g)'].apply(threshold_scale) * 2
df_scaled['Carbohydrt_(g)'] = df_scaled['Carbohydrt_(g)'] * 3
df_scaled['Calcium_(mg)'] = df_scaled['Calcium_(mg)'] * 3
df_scaled['Fiber_TD_(g)'] = df_scaled['Fiber_TD_(g)'] * 3
#df_scaled['Vit_A_RAE'] = df_scaled['Vit_A_RAE'] * 1.5
```

# KNN clustering

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 5)
model.fit(df_scaled)
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(random_state=1, perplexity=30)
tsne_repr=tsne.fit_transform(scaled_df)
```

```python
plt.scatter(tsne_repr[:,0],tsne_repr[:,1], c = model.labels_, cmap = 'viridis')
```

# AgglomerativeClustering

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
cluster.fit(scaled_df)
```

```python
plt.scatter(tsne_repr[:,0],tsne_repr[:,1], c = cluster.labels_, cmap = 'viridis')
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN

db_cluster = DBSCAN(algorithm = 'brute',eps=1 ,min_samples=10)
db_cluster.fit(scaled_df)
```

```python
plt.scatter(tsne_repr[:,0],tsne_repr[:,1], c = db_cluster.labels_, cmap = 'viridis')
```
