```python
# data
vars_ = ['Carbohydrt_(g)', 'Energ_Kcal', 'Protein_(g)', 'Lipid_Tot_(g)', 'Water_(g)', 'Fiber_TD_(g)']
```

```python
data_x = data[vars_]
data_y = data['label']
```

```python
train_x, test_x, train_y, test_y = train_test_split(data_x, data_y, test_size = 0.3)

train_x = train_x.to_numpy()
test_x = test_x.to_numpy()
train_y = train_y.to_numpy()
test_y = test_y.to_numpy()


test_y = to_categorical(test_y)
train_y = to_categorical(train_y)
```

```python
model = keras.models.Sequential([
    keras.layers.Dense(256, activation="relu"),
    keras.layers.Dropout(0.4),
    keras.layers.Dense(128, activation="relu"),
    keras.layers.Dense(5, activation="softmax")])
```

```python
model.compile(optimizer = keras.optimizers.Adam(learning_rate=0.01),
              loss = "categorical_crossentropy",
              metrics = ['accuracy'])
```

```python
learning_rate_reduction= keras.callbacks.ReduceLROnPlateau(monitor='val_acc', 
                                            patience=4, 
                                            verbose=1, 
                                            factor=0.6, 
                                            min_lr=0.00002,
                                            mode='max')
mcp_save = keras.callbacks.ModelCheckpoint('.mdl_wts.hdf5', save_best_only=True, monitor='val_acc', verbose = 1, mode='max')
```

```python
model.fit(train_x,
          train_y,
          epochs = 100,
          validation_data=(test_x,test_y),
          callbacks=[learning_rate_reduction, mcp_save]
         )
```

```python
model.load_weights('.mdl_wts.hdf5')
model.evaluate(test_x, test_y, verbose=2)
```
